# 傅里叶变换

## 1.傅里叶变换的理论基础

- 傅里叶变换：时域 <--> 频域
- 理论基础
    - 任何连续的周期信号，可以由适当的正弦曲线组合而成
    - 而任何的正弦曲线，可以由频率和振幅两个参数表示
    - 因此，任何连续的周期信号，都可以转换成频率和振幅来表示
- 傅里叶变换是可逆的
    - 可以从时域导出频域表示
    - 也可以从频域反推出（逆运算）时域表示
- 相位：时间差
    - 不是同时开始的一组正弦函数，在叠加时要体现开始的时间
    - 用 Csin(ax+b)
        - 其中a为频率, b为相位，C为振幅

## 2. numpy实现傅里叶变换

- 实现傅里叶变换的函数
    - np.fft.fft2: 实现傅里叶变换，返回一个复数数组
    - np.fft.fftshift: 将0频率（低频）分量移到频谱中心
    - 20 * np.log(np.abs(fshift)): 设置频谱的范围[0, 255]
        - 傅里叶变换得到的是一个复数数组，通过这个变换将其映射到0-255的区间之内
- 注意事项
    - 傅里叶变换的目的是得到低频、高频信息，针对低频、高频处理能够实现不同的目标
        - 左上角一般是低频，通常把它挪到中心位置（np.fft.fftshift(f)）
    - 傅里叶变换过程是可逆的，图像经过傅里叶变换、逆傅里叶变换后，能够恢复到原图像
    - 在频域对图像进行处理，会反映在逆变换图像上，就达到了图像处理的目的

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image/lena.bmp', 0)

# 傅里叶变换
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
result = 20 * np.log(np.abs(fshift))
# print(f)
# print('/*/'*30)
# print(fshift)
print(result)

# 可视化
plt.subplot(121)
plt.imshow(img, cmap=plt.cm.gray)
plt.title('original')
plt.axis('off')

plt.subplot(122)
plt.imshow(result, cmap='gray')
plt.title('result')
plt.axis('off')
```

## 2. numpy实现逆傅里叶变换

- 原理
    - 傅里叶变换：通过原始图像经过变换后得到频谱图像；
    - 逆傅里叶变换：通过频谱图像经过逆变换得到原始图像
    - 傅里叶变换的目的是在频域对图像进行处理，并通过逆变换再将处理的结果反映在原始图像上
- 逆变换函数
    - np.fft.ifft2()：实现傅里叶逆变换，返回一个复数数组
    - np.fft.ifftshift(): fft.fftshift的逆函数，将频谱图像的低频由中心位置移回到左上角
    - iimg = np.abs(逆傅里叶变换的结果) 将拟傅里叶变换的结果（复数数组）映射到0-255的灰度区间

```python
img = cv2.imread('image/boat.bmp', 0)

# 傅里叶变换
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
result = 20 * np.log(np.abs(fshift))
# print(f)
# print('/*/'*30)
# print(fshift)
# print(result)

# 逆傅里叶变换
ishift = np.fft.ifftshift(fshift)
iimg = np.fft.ifft2(ishift)
iimg = np.abs(iimg)

# 可视化
plt.figure(figsize=(16, 16))
plt.subplot(221)
plt.imshow(img, cmap=plt.cm.gray)
plt.title('original')
plt.axis('off')

plt.subplot(222)
plt.imshow(result, cmap='gray')
plt.title('result')
plt.axis('off')

plt.subplot(223)
plt.imshow(iimg, cmap='gray')
plt.title('iimg')
plt.axis('off')
```

## 4. 在频域（傅里叶域）进行高通滤波

- 低频、高频
    - 低频对应图像内变化缓慢的灰度分量。
        - 如一个草原图像中，低频对应大片颜色趋于一致的草原
    - 高频对应图像内变化越来越快的灰度分量，是由灰度的尖锐过渡造成的
        - 如一个草原图像中，狮子等动物的边缘等信息
- 滤波：分为低通和高通滤波
    - 接受（通过）或者拒绝一定频率的分量
    - 通过低频的滤波器称为低通滤波器
        - 衰减高频而通过低频，将模糊一幅图像
    - 通过高频的滤波器称为高通滤波器
        - 衰减低频而通过高频，将增强尖锐的细节，但是会导致图像的对比度降低        
- 频域滤波
    - 修改傅里叶变换的结果（频域频谱中的低频或高频分量）以达到相应的目标，然后计算IDFT返回到图像域
    - 目标：图像增强、图像去躁、边缘检测、特征提取、压缩、加密等
- 高通滤波实现：在频谱图像中去掉低频
    - rows, cols = img.shape
    - crow, ccol = int(rows/2), int(cols/2)     # 求频谱图像的中心点
    - fshift[crow-30:crow+30, ccol-30:ccol+30] = 0    # 以中心点为中心，划定一片低频区域，将其像素值转换成0

```python
# 高通滤波器实现
img = cv2.imread('image/boat.bmp', 0)

# 傅里叶变换
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)

# 高通滤波
rows, cols = img.shape
crow, ccol = int(rows/2), int(cols/2)
fshift[crow-30:crow+30, ccol-30:ccol+30] = 0

# 逆傅里叶变换
ishift = np.fft.ifftshift(fshift)
iimg = np.fft.ifft2(ishift)
iimg = np.abs(iimg)

# 可视化
plt.figure(figsize=(10, 6))
plt.subplot(121)
plt.imshow(img, cmap=plt.cm.gray)
plt.title('original')
plt.axis('off')

plt.subplot(122)
plt.imshow(iimg, cmap='gray')
plt.title('result')
plt.axis('off')
```

## 5. opencv实现傅里叶变换

- dft = cv2.dft(原始图像，转换标识)
    - 返回结果：双通道的，第一个通道是结果的实数部分，第二个通道是结果的虚数部分
    - 原始图像：输入图像要先转换为np.float32格式（np.float32(img)）
        - 一般的图像是8位二进制（256位）
    - 转换标识：一般使用flags = cv2.DFT_COMPLEX_OUTPUT,输出一个复数阵列
- dftshift = np.fft.fftshift: 将频谱图像中的0频率分量转换到频谱中心
- dst = cv2.magnitude(参数1，参数2）: 计算幅值
    - 参数1：浮点型X坐标轴，也就是实部，= dftshift[:,:,0]
    - 参数2：浮点型Y坐标轴，也就是虚部, = dftshift[:,:,1]
    - dst(I) = sqrt(x^2 + y^2)
- 20 * np.log(幅值)： 将复数转换到0-255区间

```python
img = cv2.imread('image/lena.bmp', 0)

# 傅里叶变换
dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
dftshift = np.fft.fftshift(dft)
result = 20 * np.log(cv2.magnitude(dftshift[:,:,0], dftshift[:,:,1]))

# 可视化
plt.figure(figsize=(10, 6))
plt.subplot(121)
plt.imshow(img, cmap=plt.cm.gray)
plt.title('original')
plt.axis('off')

plt.subplot(122)
plt.imshow(result, cmap='gray')
plt.title('result')
plt.axis('off')
```

## 6. opencv实现逆傅里叶变换


- np.fft.ifftshift()：将低频由频谱的中心移到左上角
    - np.fft.fftshift()的逆函数
- 返回结果 = cv2.idft(原始数据)
    - 返回结果：取决于原始数据的类型和大小，得到是的双通道（有实部和虚部）
    - 原始数据：实数或者复数均可
- 返回值 = cv2.magnitude(参数1，参数2) 
    - 参数同傅里叶变换

```python
img = cv2.imread('image/lena.bmp', 0)

# 傅里叶变换
dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
dftshift = np.fft.fftshift(dft)
# result = 20 * np.log(cv2.magnitude(dftshift[:,:,0], dftshift[:,:,1]))

# 逆傅里叶变换
idftshift = np.fft.ifftshift(dftshift)
iimg = cv2.idft(idftshift)
iimg = cv2.magnitude(iimg[:,:,0], iimg[:,:,1])

# 可视化
plt.figure(figsize=(10, 6))
plt.subplot(121)
plt.imshow(img, cmap=plt.cm.gray)
plt.title('original')
plt.axis('off')

plt.subplot(122)
plt.imshow(iimg, cmap='gray')
plt.title('result')
plt.axis('off')
```

## 7. 在频域（傅里叶域）进行低通滤波

- 实现原理：
    - 构造一个掩膜图像(中间部分1，其余部分0)
    - 频谱图像乘以掩膜图像，得到滤波结果图像
- 生成低通滤波结果图像的代码
    - rows, cols = img.shape
    - crow,ccol = int(rows/2), int(cols/2)
    - mask = np.zeros((rows, cols, 2), np.uint8)
    - mask[crow-30:crow+30, ccol-30:ccol+30] = 1

```python
img = cv2.imread('image/lena.bmp', 0)

# 傅里叶变换
dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
dftshift = np.fft.fftshift(dft)  # 频谱图像

# 构造低通滤波器
rows, cols = img.shape
crow,ccol = int(rows/2), int(cols/2)
mask = np.zeros((rows, cols, 2), np.uint8)
mask[crow-50:crow+50, ccol-50:ccol+50] = 1

# 滤波，生成滤波结果图像
fshift = dftshift * mask

# 逆傅里叶变换
idftshift = np.fft.ifftshift(fshift)
iimg = cv2.idft(idftshift)
iimg = cv2.magnitude(iimg[:,:,0], iimg[:,:,1])

# 可视化
plt.figure(figsize=(10, 6))
plt.subplot(121)
plt.imshow(img, cmap=plt.cm.gray)
plt.title('original')
plt.axis('off')

plt.subplot(122)
plt.imshow(iimg, cmap='gray')
plt.title('result')
plt.axis('off')
```
